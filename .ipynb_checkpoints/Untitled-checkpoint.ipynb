```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('../Computer-Vision-with-Python/DATA/00-puppy.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
img2 = cv2.imread('../Computer-Vision-with-Python/DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
img1.shape
```

```python
img2.shape
```

```python
#BLENDING IMAGES OF THE SAME SIZE
img1 = cv2.resize(img1, (1200,1200))
img2 = cv2.resize(img2, (1200,1200))
blended = cv2.addWeighted(src1=img1, alpha=0.9, src2=img2, beta=0.5, gamma=0)
plt.imshow(blended)
```

```python
#OVERLAY SMALL IMAGE ON TOP OF A LARGER IMAGE (NO BLENDING)
#Numpy Reassignment
img1 = cv2.imread('../Computer-Vision-with-Python/DATA/00-puppy.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread('../Computer-Vision-with-Python/DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
img2 = cv2.resize(img2,(600,600))
x_offset = 0
y_offset = 0
x_end = x_offset + img2.shape[1]
y_end = y_offset + img2.shape[0]
img1[y_offset:y_end,x_offset:x_end] = img2
plt.imshow(img1)
```

```python
#BLEND TOGETHER IMAGES OF DIFFERENT SIZES
img1 = cv2.imread('../Computer-Vision-with-Python/DATA/00-puppy.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img1 = cv2.flip(img1,1)
img2 = cv2.imread('../Computer-Vision-with-Python/DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
img2 = cv2.resize(img2,(600,600))

rows1,cols1,channels1 = img1.shape
rows2,cols2,channels2 = img2.shape
x_offset = cols1 - cols2
y_offset = rows1 - rows2
roi = img1[y_offset:rows1,x_offset:cols1] #region of interest
plt.imshow(roi)
```

```python
img2gray = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
plt.imshow(img2gray)
```

```python
plt.imshow(img2gray,cmap="gray")
```

```python
mask_inv = cv2.bitwise_not(img2gray)
plt.imshow(mask_inv,cmap="gray")
```

```python
mask_inv.shape
```

```python
import numpy as np
white_background = np.full(img2.shape,255,dtype=np.uint8)
bk = cv2.bitwise_or(white_background,white_background,mask=mask_inv)
bk.shape
plt.imshow(bk)
```

```python
foreground = cv2.bitwise_or(img2,img2,mask=mask_inv)
plt.imshow(foreground)
```

```python
final_roi = cv2.bitwise_or(roi,foreground)
plt.imshow(final_roi)
```

```python
x_end = x_offset + final_roi.shape[1]
y_end = y_offset + final_roi.shape[0]
img1[y_offset:y_end,x_offset:x_end] = final_roi
plt.imshow(img1)
```
